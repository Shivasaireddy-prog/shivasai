```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
df = pd.read_csv(url)

# Select a feature to normalize, for example, 'rm' (average number of rooms per dwelling)
feature = 'rm'

# Compute MinMax value
min_value = df[feature].min()
max_value = df[feature].max()
df[f'{feature}_minmax'] = (df[feature] - min_value) / (max_value - min_value)

# Print MinMax values
print(f"Min value of {feature}: {min_value}")
print(f"Max value of {feature}: {max_value}")
print(f"MinMax normalized values of {feature}:\n{df[f'{feature}_minmax'].head()}")

# Plot the distribution of the original feature
plt.figure(figsize=(10, 6))
sns.histplot(df[feature], kde=True, color='blue')
plt.title(f'Distribution of Original {feature}')
plt.xlabel(feature)
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of the normalized feature
plt.figure(figsize=(10, 6))
sns.histplot(df[f'{feature}_minmax'], kde=True, color='green')
plt.title(f'Distribution of MinMax Normalized {feature}')
plt.xlabel(f'{feature}_minmax')
plt.ylabel('Frequency')
plt.show()
```
